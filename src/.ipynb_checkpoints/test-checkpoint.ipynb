### 从v3文件中提取路线

```python
from osgeo import ogr
import cv2
```

```python
driver = ogr.GetDriverByName("ESRI Shapefile") 

filename = "../res/v3/Road_vector/Road.shp"

dataSource = driver.Open(filename, 0)

layer = dataSource.GetLayer(0)
```

```python
def coordinate_transfer(old_x,old_y,layer_shape, sketch_shape):
    width = layer_shape[1]-layer_shape[0]
    height = layer_shape[3]-layer_shape[2]
    new_x = round((old_x-layer_shape[0])/width*sketch_shape[1])
    new_y = round((old_y-layer_shape[2])/height*sketch_shape[0])
    return new_x, new_y
```

```python
def road_extract(layer, sketch_shape):
    line_list=[]
    layer_shape = layer.GetExtent()
    for i in range(0,layer.GetFeatureCount()):
        feat = layer.GetFeature(i)
        geom = feat.geometry()
        line=[]
        if geom.GetGeometryCount()>0:
            for j in range(0, geom.GetGeometryCount()):
                g = feat.geometry().GetGeometryRef(j)
                for p in range(0, g.GetPointCount()):
                    pt=g.GetPoint(p)
                    new_x,new_y=coordinate_transfer(pt[0],pt[1],layer_shape, sketch_shape)
                    line.append((new_x,new_y))
        else:
            for p in range(0, geom.GetPointCount()):
                pt=geom.GetPoint(p)
                new_x,new_y=coordinate_transfer(pt[0],pt[1],layer_shape, sketch_shape)
                line.append((new_x,new_y))
        line_list.append(line)
    return line_list
```

```python
tif = TIFF.open("../res/v3/Label_1m/Label_1m.tif", mode='r')  # 打開tiff文件進行讀取
```

```python
im = tif.read_image()
```

```python
im.shape
```

```python
plt.imshow(im)
```

```python
roads=road_extract(layer, im.shape)
```

```python
back_ground = np.zeros((im.shape[0],im.shape[1],3))
```

```python
for road in roads:
    p1 = road[0]
    for index, p in enumerate(road):
            if index == 0:
                continue
            p2 = p
            back_ground=cv2.line(back_ground, p1, p2, (255, 0, 0), 40)
            p1 = p
```
