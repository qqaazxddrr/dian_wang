```python
def coordinate_transfer(old_x,old_y,layer_shape, sketch_shape):
    width = layer_shape[1]-layer_shape[0]
    height = layer_shape[3]-layer_shape[2]
    new_x = round((old_x-layer_shape[0])/width*sketch_shape[0])
    new_y = round((old_y-layer_shape[2])/height*sketch_shape[1])
    return new_x, new_y
```

```python
def road_extract(layer):
    line_list=[]
    layer_shape = layer.GetExtent()
    for i in range(0,layer.GetFeatureCount()):
        feat = layer.GetFeature(i)
        geom = feat.geometry()
        if geom is None:
            continue
        line=[]
        if geom.GetGeometryCount()>0:
            for j in range(0, geom.GetGeometryCount()):
                g = feat.geometry().GetGeometryRef(j)
                for p in range(0, g.GetPointCount()):
                    pt=g.GetPoint(p)
#                     new_x,new_y=coordinate_transfer(pt[0],pt[1],layer_shape, sketch_shape)
#                     line.append((new_x,new_y))
                    line.append((round(pt[0]),8100-round(pt[1])))
        else:
            for p in range(0, geom.GetPointCount()):
                pt=geom.GetPoint(p)
#                 new_x,new_y=coordinate_transfer(pt[0],pt[1],layer_shape, sketch_shape)
#                 line.append((new_x,new_y))
                line.append((round(pt[0]),8100-round(pt[1])))
        line_list.append(line)
    return line_list
```

```python
def getSRSPair(prj):
    '''
    获得给定数据的投影参考系和地理参考系
    :param dataset: GDAL地理数据
    :return: 投影参考系和地理参考系
    '''
    prosrs = osr.SpatialReference()
    prosrs.ImportFromWkt(prj)
    geosrs = prosrs.CloneGeogCS()
    return prosrs, geosrs
```

```python
def geo2lonlat(prj, x, y):
    '''
    将投影坐标转为经纬度坐标（具体的投影坐标系由给定数据确定）
    :param dataset: GDAL地理数据
    :param x: 投影坐标x
    :param y: 投影坐标y
    :return: 投影坐标(x, y)对应的经纬度坐标(lon, lat)
    '''
    prosrs, geosrs = getSRSPair(prj)
    ct = osr.CoordinateTransformation(prosrs, geosrs)
    coords = ct.TransformPoint(x, y)
    return coords[:2]
```

## 小测试，用极坐标采样邻居节点

```python
def neighbors_generator(radius_inner, radius_outer, sample_n):
    quadrant = []
    result = set()
    for x in range(0, radius_outer + 1):
        for y in range(0, radius_outer + 1):
            if radius_inner ** 2 <= x ** 2 + y ** 2 <= radius_outer ** 2:
                if x % sample_n == 0 and y % sample_n == 0:
                    quadrant.append((x, y))

    for x, y in quadrant:
        result.add((x, y))
        result.add((-x, y))
        result.add((x, -y))
        result.add((-x, -y))
    return list(result)
```

```python
def neighbors_generator(radius_inner, radius_outer, sample_n):
    result=set()
    for _ in range(sample_n):
        degree = random.uniform(0,360)/180*np.pi
        length = random.uniform(radius_inner,radius_outer)
        x=round(np.cos(degree)*length)
        y=round(np.sin(degree)*length)
        result.add((x,y))
    return list(result)
```

```python
def neighbors_generator_v2(radius_inner, radius_outer, length_part, degree_delta):
    result = set()
    length_delta = int((radius_outer - radius_inner)/length_part)
    lengths = []
    for i in range(length_part):
        lengths.append(i*length_delta+radius_inner)
    # lengths = [radius_inner, radius_outer, int((radius_inner+radius_outer)/2)]
    for degree in range(0,360,degree_delta):
        for length in lengths:
            x = int(np.cos(degree/180*np.pi) * length)
            y = int(np.sin(degree / 180 * np.pi) * length)
            result.add((x,y))
    return list(result)
```

# 路线评估

```python
import math
from bresenham import bresenham
import numpy as np
```

```python
DL = 1  # 1 道路
SX = 2  # 2 水系
FW = 3  # 3 房屋(居民用地)
JZYD = 4  # 4 建设用地
NT = 5  # 5 农田
LM = 6  # 6 林木
class1 = [5, 6]
class2 = [2]
class3 = [3, 1]
class4 = [4]
```

```python
def distance_between(n1, n2):
    """this method always returns 1, as two 'neighbors' are always adajcent"""
    (x1, y1) = n1
    (x2, y2) = n2
    return math.hypot(x2 - x1, y2 - y1)
```

```python
# 长度
def length_of_path(path):
    p0 = path[0]
    length = []
    for index, p in enumerate(path):
        if index==0:
            continue
        length.append(distance_between(p0, p))
        p0 = p
    return length
```

```python
# 长度
total_length = sum(length_of_path(path))
total_length
```

```python
# 塔基数量
num_tower = len(path)
num_tower
```

```python
# 经历各类地块数目
def type_sum(sketch, path):
    p0 = path[0]
    num1=0
    num2=0
    num3=0
    num4=0
    for index, p in enumerate(path):
        if index==0 :
            continue
        route = list(bresenham(p0[0], p0[1], p[0], p[1]))
        for item in route:
            if sketch[item[1]][item[0]] in class1:
                num1=num1+1
            if sketch[item[1]][item[0]] in class2:
                num2=num2+1
            if sketch[item[1]][item[0]] in class3:
                num3=num3+1
            if sketch[item[1]][item[0]] in class4:
                num4=num4+1
        p0 = p
    return num1, num2, num3, num4
```

```python
type_sum(sketch, path)
```

```python
# 塔基所在地类数目
def tower_type_sum(sketch, path):
    num1=0
    num2=0
    num3=0
    num4=0
    
    for tower in path:
        if sketch[tower[1]][tower[0]] in class1:
            num1=num1+1
        if sketch[tower[1]][tower[0]] in class2:
            num2=num2+1
        if sketch[tower[1]][tower[0]] in class3:
            num3=num3+1
        if sketch[tower[1]][tower[0]] in class4:
            num4=num4+1
    return num1, num2, num3, num4
```

```python
tower_type_sum(sketch, path)
```
