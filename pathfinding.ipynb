```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
gridMap = np.load('map.npy')
```

```python
plt.imshow(gridMap)
```

```python
maze = cv2.inRange(gridMap, 2.9, 3.1)
plt.imshow(maze)
```

```python
from pathfinder import pathfinder
```

```python
start = (1400,3500)
end = (2200,900)
```

```python
path = list(pathfinder(maze).astar(start, end))
```

```python
maze_viz = cv2.cvtColor(maze,cv2.COLOR_GRAY2RGB)
```

```python
p1 = path[0]
for index,p in enumerate(path):
    if index == 0:
        continue
    p2 = p
    cv2.line(maze_viz, p1, p2, (255,50,0), 50)
    p1 = p
plt.imshow(maze_viz)
```
