```python
img = cv2.imread(fp)
im = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
row, col, _ = im.shape
```

```python
row,col
```

```python
sketch = np.zeros([row, col])
```

```python
time1 = time.time()
for i in range(row):
    for j in range(col):
        tmp = im[i,j].tolist()
        if tmp == [0,0,255]:
            sketch[i,j] = 2
            continue
        if tmp == [0,255,0]:
            sketch[i,j] = 1
            continue
        if tmp==[85, 107, 47]:
            sketch[i,j] = 1
            continue
        if tmp==[210, 105, 30]:
            sketch[i,j] = 4
            continue
        if tmp==[255, 0, 0]:
            sketch[i,j] = 3
            continue
        if tmp==[255, 255, 0]:
            sketch[i,j] = 3
            continue
    if i%1000==0:
        time2 = time.time()
        print("row:{},time:{}".format(i,time2-time1))
        time1 = time2
```

```python
cv2.imwrite("sketch.png",sketch)
```

# 图像处理

```python
sketch = np.load(fp)
```

```python
sketch.shape
```

```python
new_row=round(sketch.shape[0]/3)
new_col=round(sketch.shape[1]/3)
```

```python
sampled_sketch = np.zeros([new_row, new_col])
```

```python
sample_core = np.array([[0,0],[0,1],[0,2],[1,0],[1,1],[1,2],[2,0],[2,1],[2,2]])
```

```python
anchor = np.array([0,0])
```

```python
time1 = time.time()
for i in range(new_row):
    for j in range(new_col):
        anchor=np.array([i,j])
        type_list=[]
        for unit in sample_core:
            index = anchor+unit
            type_list.append(sketch[index[0],index[1]])
            if 4 in type_list:
                sampled_sketch[anchor[0],anchor[1]]=4
                break
            elif 3 in type_list:
                sampled_sketch[anchor[0],anchor[1]]=3
                break
            elif 2 in type_list:
                sampled_sketch[anchor[0],anchor[1]]=2
                break
            else:
                sampled_sketch[anchor[0],anchor[1]]=1
        type_list=[]
    if i%1000==0:
        time2 = time.time()
        print("row:{},time:{}".format(i,time2-time1))
        time1 = time2
        
    
        
        
```
